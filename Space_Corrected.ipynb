```python
data5 = pd.DataFrame({'status_company': ['частная', 'государственная', 'частная', 'государственная']})
```

```python
for index, row in data5.iterrows():
    if row['status_company'] == 'частная':
        data5.at[index, 'status_company'] = 'private'
    elif row['status_company'] == 'государственная':
        data5.at[index, 'status_company'] = 'state'
```
